```python
# 고객별 첫 구매 Cohort 설정
df['First_Purchase_Month'] = df.groupby('CID')['Purchase_Date'].transform('min').dt.to_period('M')
```

```python
# 구매 월
df['Purchase_Month'] = df['Purchase_Date'].dt.to_period('M')
```

```python
df
```

```python
# 첫 구매후 재구매 연도 계산
df['Difference'] = (df['Purchase_Month'] - df['First_Purchase_Month']).apply(lambda x: x.n)
```

```python
cohort_counts = df.groupby(['First_Purchase_Month', 'Difference'])['CID'].nunique().unstack()
cohort_sizes = cohort_counts.iloc[:, 0]
cohort_retention = cohort_counts.divide(cohort_sizes, axis=0)
```

```python
cohort_data = df.pivot_table(index='First_Purchase_Month', 
                             columns='Difference', 
                             values='CID', 
                             aggfunc=pd.Series.nunique)

# Cohort별 첫 달 기준으로 리텐션율 계산
cohort_retention = cohort_data.div(cohort_data.iloc[:, 0], axis=0).round(3)

```
